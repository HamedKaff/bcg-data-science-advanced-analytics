## Data visualization

If you're working in Python, two of the most popular packages for visualization are `matplotlib` and `seaborn`. We highly recommend you use these, or at least be familiar with them because they are ubiquitous!

Below are some functions that you can use to get started with visualizations. 

```python
def plot_stacked_bars(dataframe, title_, size_=(18, 10), rot_=0, legend_="upper right"):
    """
    Plot stacked bars with annotations
    """
    ax = dataframe.plot(
        kind="bar",
        stacked=True,
        figsize=size_,
        rot=rot_,
        title=title_
    )

    # Annotate bars
    annotate_stacked_bars(ax, textsize=14)
    # Rename legend
    plt.legend(["Retention", "Churn"], loc=legend_)
    # Labels
    plt.ylabel("Company base (%)")
    plt.show()

def annotate_stacked_bars(ax, pad=0.99, colour="white", textsize=13):
    """
    Add value annotations to the bars
    """

    # Iterate over the plotted rectanges/bars
    for p in ax.patches:
        
        # Calculate annotation
        value = str(round(p.get_height(),1))
        # If value is 0 do not annotate
        if value == '0.0':
            continue
        ax.annotate(
            value,
            ((p.get_x()+ p.get_width()/2)*pad-0.05, (p.get_y()+p.get_height()/2)*pad),
            color=colour,
            size=textsize
        )

def plot_distribution(dataframe, column, ax, bins_=50):
    """
    Plot variable distirbution in a stacked histogram of churned or retained company
    """
    # Create a temporal dataframe with the data to be plot
    temp = pd.DataFrame({"Retention": dataframe[dataframe["churn"]==0][column],
    "Churn":dataframe[dataframe["churn"]==1][column]})
    # Plot the histogram
    temp[["Retention","Churn"]].plot(kind='hist', bins=bins_, ax=ax, stacked=True, color = ["#44a5c2", "#024b7a"])
    # X-axis label
    ax.set_xlabel(column)
    # Change the x-axis to plain style
    ax.ticklabel_format(style='plain', axis='x')
```

Thhe first function `plot_stacked_bars` is used to plot a stacked bar chart. An example of how you could use this is shown below:

```python
churn = client_df[['id', 'churn']]
churn.columns = ['Companies', 'churn']
churn_total = churn.groupby(churn['churn']).count()
churn_percentage = churn_total / churn_total.sum() * 100
plot_stacked_bars(churn_percentage.transpose(), "Churning status", (5, 5), legend_="lower right")
```

# <font color='navy'>5. Hypothesis Evaluation</font>
 

 


Now that we have performed a bit of analysis and data exploration, and have gained an understanding of the data, we cat start investigating the hypothesis and see if there is a correlation between prices and customer churn. First, in order to do that, we need to define and specify what price sensitivity is. 

    > Since we have the consumption data for each of the companies for the year of 2015, we will create new         features to measure the price sensitivity using the average of the year, the last 6 months and the last 3       months. Note that we sat aa minimum of 3 months past since that the definition of the churn field is `churn: has the client churned over the next 3 months`



```python
# Transform date columns to datetime type
client_df["date_activ"] = pd.to_datetime(client_df["date_activ"], format='%Y-%m-%d')
client_df["date_end"] = pd.to_datetime(client_df["date_end"], format='%Y-%m-%d')
client_df["date_modif_prod"] = pd.to_datetime(client_df["date_modif_prod"], format='%Y-%m-%d')
client_df["date_renewal"] = pd.to_datetime(client_df["date_renewal"], format='%Y-%m-%d')
price_df['price_date'] = pd.to_datetime(price_df['price_date'], format='%Y-%m-%d')
```

```python


# Create mean average data
mean_year = price_df.groupby(['id']).mean().reset_index()
mean_6m = price_df[price_df['price_date'] > '2015-06-01'].groupby(['id']).mean().reset_index()
mean_3m = price_df[price_df['price_date'] > '2015-10-01'].groupby(['id']).mean().reset_index()



mean_year = mean_year.rename(index=str, columns={"price_off_peak_var": "mean_year_price_p1_var", 
                                                 "price_peak_var": "mean_year_price_p2_var", 
                                                 "price_mid_peak_var": "mean_year_price_p3_var", 
                                                 "price_off_peak_fix": "mean_year_price_p1_fix", 
                                                 "price_peak_fix": "mean_year_price_p2_fix", 
                                                 "price_mid_peak_fix": "mean_year_price_p3_fix"})

mean_year["mean_year_price_off_peak"] = mean_year["mean_year_price_p1_var"] + mean_year["mean_year_price_p1_fix"]
mean_year["mean_year_price_peak"] = mean_year["mean_year_price_p2_var"] + mean_year["mean_year_price_p2_fix"]
mean_year["mean_year_price_mid_peak"] = mean_year["mean_year_price_p3_fix"] + mean_year["mean_year_price_p3_var"]




mean_6m = mean_6m.rename(index=str, columns={"price_off_peak_var": "mean_6m_price_p1_var", 
                                                 "price_peak_var": "mean_6m_price_p2_var", 
                                                 "price_mid_peak_var": "mean_6m_price_p3_var", 
                                                 "price_off_peak_fix": "mean_6m_price_p1_fix", 
                                                 "price_peak_fix": "mean_6m_price_p2_fix", 
                                                 "price_mid_peak_fix": "mean_6m_price_p3_fix"})

mean_6m["mean_6m_price_off_peak"] = mean_6m["mean_6m_price_p1_var"] + mean_6m["mean_6m_price_p1_fix"]
mean_6m["mean_6m_price_peak"] = mean_6m["mean_6m_price_p2_var"] + mean_6m["mean_6m_price_p2_fix"]
mean_6m["mean_6m_price_mid_peak"] = mean_6m["mean_6m_price_p3_fix"] + mean_6m["mean_6m_price_p3_var"]



mean_3m = mean_3m.rename(index=str, columns={"price_off_peak_var": "mean_3m_price_p1_var", 
                                                 "price_peak_var": "mean_3m_price_p2_var", 
                                                 "price_mid_peak_var": "mean_3m_price_p3_var", 
                                                 "price_off_peak_fix": "mean_3m_price_p1_fix", 
                                                 "price_peak_fix": "mean_3m_price_p2_fix", 
                                                 "price_mid_peak_fix": "mean_3m_price_p3_fix"})

mean_3m["mean_3m_price_off_peak"] = mean_3m["mean_3m_price_p1_var"] + mean_3m["mean_3m_price_p1_fix"]
mean_3m["mean_3m_price_peak"] = mean_3m["mean_3m_price_p2_var"] + mean_3m["mean_3m_price_p2_fix"]
mean_3m["mean_3m_price_mid_peak"] = mean_3m["mean_3m_price_p3_fix"] + mean_3m["mean_3m_price_p3_var"]



# Merge into 1 dataframe
price_features = pd.merge(mean_year, mean_6m, on='id')
price_features = pd.merge(price_features, mean_3m, on='id')
```

```python
price_analysis = pd.merge(price_features, client_df[['id', 'churn']], on='id')
corr = price_analysis.corr()
# Plot correlation
plt.figure(figsize=(20,18))
sns.heatmap(corr, xticklabels=corr.columns.values, yticklabels=corr.columns.values, annot = True, cmap="YlGnBu",annot_kws={'size':10})
# Axis ticks size
plt.xticks(fontsize=13.7)
plt.yticks(fontsize=13.7)
plt.show()
```

As the heatmap shows, there is an extremely weak correlation between customer churn and prices. To simplify reading the heatmap, from the very bottom right, reading the values vertically, and the result is that all values are almost close to 0. As we know, correlation values that are close to 0 mean there is no correlation, but the closer the values to 1 or -1, the stronger the correlation. 

Therefore, it is important to perform feature engineering, perform deeper analysis to find the relationship between churn and prices, and eventually create a model to predict the churn.
